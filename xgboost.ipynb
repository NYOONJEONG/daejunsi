```python
g1train_grouped = g1train.groupby('날짜').mean()
```

```python
g1train_grouped
```

```python
g1test_grouped = g1test.groupby('날짜').mean()
g1test_grouped
```

```python
from sklearn.model_selection import train_test_split
X_g1_train, X_g1_val, y_g1_train, y_g1_val = train_test_split(X_train1,g1train_grouped['이용자 수'],test_size=0.2, random_state=42)
```

```python
X_g1_train = X_g1_train.sort_values(by='날짜')
X_g1_train = X_g1_train.drop(['날짜'], axis=1)
```

```python
X_g1_val = X_g1_val.sort_values(by='날짜')
X_g1_val = X_g1_val.drop(['날짜'], axis=1)
```

```python
X_test1 = X_test1.drop(['날짜'],axis=1)
```

```python
import xgboost as xgb
from xgboost import XGBRegressor
```

```python
xgb_model = xgb.XGBRegressor(n_estimators=500, learning_rate= 0.1, max_depth=4,random_state=42)

xgb_model.fit(X_g1_train, y_g1_train,
              eval_set=[(X_g1_val, y_g1_val)], eval_metric = 'rmse', early_stopping_rounds=100)

xgb_pred = xgb_model.predict(X_test1)
```

```python
from sklearn.metrics import mean_squared_error
def rmse(y, pred):
    return np.sqrt(mean_squared_error(y,pred))
```

```python
g1test
```

```python
rmse(g1test_grouped['이용자 수'], xgb_pred)
```

Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV, KFold
```

```python
xgb_param_grid = {
    'n_estimators' : [100,200,300,400,500,600],
    'learning_rate' : [0.01,0.05,0.1,0.15,0.2],
    'max_depth': [4,6,8,10,12]
}

cv = KFold(n_splits=6, random_state=1)

fit_parameters = {'eval_set' : [(X_g1_val, y_g1_val)], 'eval_metric' : 'rmse'}

xgb1 = XGBRegressor()

xgb_grid = GridSearchCV(xgb1, param_grid = xgb_param_grid, cv=cv, n_jobs= -1, scoring='neg_mean_squared_error', verbose = True)
```

```python
xgb_grid.fit(X_g1_train, y_g1_train)

print(xgb_grid.best_score_)
print(xgb_grid.best_params_)
```

```python
cvres = xgb_grid.cv_results_ 
for mean_score, params in zip(cvres["mean_test_score"], cvres["params"]): print(np.sqrt(-mean_score), params)
```
